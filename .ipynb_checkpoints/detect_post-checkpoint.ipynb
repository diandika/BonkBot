```python
import requests
import unidecode
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
page = requests.get('https://notes.qoo-app.com/en')
page_source = BeautifulSoup(page.text, "html.parser")
page_source
```

```python
posts_list = page_source.findAll("div", class_="qoo-note-view")
posts_list
```

```python
def normalize(s):
    new_s = s.lower()
    new_s = unidecode.unidecode(new_s)
    new_s = new_s.replace("4", "a")
    new_s = new_s.replace("3", "e")
    
    return new_s
```

```python
post_info_list = []
for post in posts_list:
    post_info = {"user": "", "title": "", "desc": ""}
    
    post_info["user"] = post.findAll("a", class_="artist")[0]["href"]
    title = post.findAll("strong", class_="content-title")
    if len(title) > 0 :
        post_info["title"] = normalize(title[0].text)
    desc = post.findAll("cite", class_="description")
    if len(desc) > 0 :
        post_info["desc"] = normalize(desc[0].text.strip())
    
    post_info_list.append(post_info)
```

```python
for post in post_info_list:
    post["title"] = normalize(post["title"])
    print(post["title"])
    post["desc"] = normalize(post["desc"])
    print(post["desc"])
```
